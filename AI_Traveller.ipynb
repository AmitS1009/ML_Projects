```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Collecting the data :

```python
df = pd.read_csv("/content/Data4.csv")
```

Analysis of Data :

```python
df.shape
```

```python
df.head()
```

```python

print(df.info())
print(df.describe())
```

Preprocessing the data :

```python
df = df.drop(columns=["Name"])
```

```python
df.head(1)
```

```python
# Converting 'Preferences' column into numeric codes
df.columns = df.columns.str.strip()

df['Preferences'] = df['Preferences'].astype('category').cat.codes
print(df)
```

```python
df.head(2)
```

```python
df['Destination'] = df['Destination'].astype('category').cat.codes
df.head(10)
```

Splitting the data :

```python
X = df.drop(columns=["Destination"])
Y = df["Destination"]
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.head(3)
```

```python
y_train.head(4)
```

Model Building :

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
input_data = (30, 40000, "Adventure", 7)

# Map preferences to numbers (as used during training)
preferences_mapping = {
    "Adventure": 1,
    "Relaxation": 2,
    "Culture": 3,
    "Nature": 4,
    "Spirituality": 5
}

# Convert categorical preference to numerical value
input_data_processed = (
    input_data[0],  # Age
    input_data[1],  # Budget
    preferences_mapping[input_data[2]],  # Preferences
    input_data[3]   # Duration
)

# Changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data_processed)

# Reshape the data as we are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Predict destination
prediction = model.predict(input_data_reshaped)

if (predictions[0]==0):
  print('Predicted Destination: Darjeeling')
elif(predictions[0]==1):
  print('Predicted Destination: Gangtok')
elif(predictions[0]==2):
  print('Predicted Destination: Tawang')
```
